```python
import pandas as pd
```

```python
df_bike = pd.read_csv('model_1/2022년_일별_행정구_대여량_총계.csv')
df_bike.drop('Unnamed: 0',axis=1,inplace=True)
df_human = pd.read_csv('model_1/2022년_시간대구분있는행정구인구수.csv') # 시간대 평균으로 groupby
df_human.drop('Unnamed: 0',axis=1,inplace=True)
df_bike['월']=df_bike['월'].astype(int)
df_bike['일']=df_bike['일'].astype(int)
def date_sep(row) :
    row = row.copy()
    row['월'] = int(str(row['기준일ID'])[4:6])
    row['일'] = int(str(row['기준일ID'])[6:])
    return row
df_human = df_human.apply(date_sep,axis=1)
df_human.drop('기준일ID',axis=1, inplace=True)
df_no_weather = df_bike.merge(df_human,on = ['행정구','월','일'])
df_no_weather.to_csv('model_1/날짜붙이면2022년데이터완성.csv',index=False)
```
