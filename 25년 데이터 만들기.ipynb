```python
import pandas as pd
```

```python
weather = pd.read_csv('data/22년부터일일단위기상.csv', encoding='cp949',parse_dates=['일시'])
weather2025 = weather[weather['일시']>'2024-12-31']
weather2025.fillna(0)
```

```python
df_1 = pd.read_csv(r"C:\Users\wjdwh\workspace\project\--PM----\model_1\2025데이터\LOCAL_PEOPLE_DONG_202501.csv",index_col=False)[['기준일ID','시간대구분','행정동코드','총생활인구수']]
df_1
```

```python
df_1 = df_1.groupby(['기준일ID','행정동코드'],as_index=False)['총생활인구수'].median()
```

```python
def row_date(row) :
    row = row.copy()
    row['일시'] = (str(row['기준일ID'].astype('int'))[:4] + '-' + str(row['기준일ID'].astype('int'))[4:6] + '-' + str(row['기준일ID'].astype('int'))[6:])
    return row
```

```python
row_date(df_1.iloc[0])
```

```python
df_1 = df_1.apply(row_date,axis=1)
```

```python
df_1.drop('기준일ID',axis=1, inplace=True)
```

```python
df_1['행정동코드']=df_1['행정동코드'].astype('int')
```

```python
code = pd.read_excel('model_1/행정동코드_매핑정보_20241218.xlsx')
code = code[['행자부행정동코드','시군구명']]
code.drop(0,inplace=True)
```

```python
code.columns=['행정동코드','행정구']
```

```python
# code.info() # 행정동코드 object
# df_1.info() #  행정동코드 int32
code['행정동코드'] = code['행정동코드'].astype('int')
df_1 = df_1.merge(code,on=['행정동코드'],how='outer')
```

```python
df_2 = pd.read_csv(r"C:\Users\wjdwh\workspace\project\--PM----\model_1\2025데이터\LOCAL_PEOPLE_DONG_202502.csv",index_col=False)[['기준일ID','시간대구분','행정동코드','총생활인구수']]

df_2 = df_2.groupby(['기준일ID','행정동코드'],as_index=False)['총생활인구수'].median()

df_2 = df_2.apply(row_date,axis=1)

df_2.drop('기준일ID',axis=1, inplace=True)

df_2['행정동코드']=df_2['행정동코드'].astype('int')
code['행정동코드'] = code['행정동코드'].astype('int')
df_2 = df_2.merge(code,on=['행정동코드'],how='outer')
df_2.isna().sum()
```

```python
df_3 = pd.read_csv(r"C:\Users\wjdwh\workspace\project\--PM----\model_1\2025데이터\LOCAL_PEOPLE_DONG_202503.csv",index_col=False)[['기준일ID','시간대구분','행정동코드','총생활인구수']]

df_3 = df_3.groupby(['기준일ID','행정동코드'],as_index=False)['총생활인구수'].median()

df_3 = df_3.apply(row_date,axis=1)

df_3.drop('기준일ID',axis=1, inplace=True)

df_3['행정동코드']=df_3['행정동코드'].astype('int')
code['행정동코드'] = code['행정동코드'].astype('int')
df_3 = df_3.merge(code,on=['행정동코드'],how='outer')
df_3.isna().sum()
```

```python
df_4 = pd.read_csv(r"C:\Users\wjdwh\workspace\project\--PM----\model_1\2025데이터\LOCAL_PEOPLE_DONG_202504.csv",index_col=False)[['기준일ID','시간대구분','행정동코드','총생활인구수']]

df_4 = df_4.groupby(['기준일ID','행정동코드'],as_index=False)['총생활인구수'].median()

df_4 = df_4.apply(row_date,axis=1)

df_4.drop('기준일ID',axis=1, inplace=True)

df_4['행정동코드']=df_4['행정동코드'].astype('int')
code['행정동코드'] = code['행정동코드'].astype('int')
df_4 = df_4.merge(code,on=['행정동코드'],how='outer')
df_4.isna().sum()
```

```python
df_5 = pd.read_csv(r"C:\Users\wjdwh\workspace\project\--PM----\model_1\2025데이터\LOCAL_PEOPLE_DONG_202505.csv",index_col=False)[['기준일ID','시간대구분','행정동코드','총생활인구수']]
df_5 = df_5.groupby(['기준일ID','행정동코드'],as_index=False)['총생활인구수'].median()

df_5 = df_5.apply(row_date,axis=1)

df_5.drop('기준일ID',axis=1, inplace=True)

df_5['행정동코드']=df_5['행정동코드'].astype('int')
code['행정동코드'] = code['행정동코드'].astype('int')
df_5 = df_5.merge(code,on=['행정동코드'],how='outer')
df_5.isna().sum()
```

```python
df_past = pd.read_csv('model_1/total_data.csv')[['일시','행정구','총대여량']]
# df_final.info() # 일시 datetime64[ns] 행정구 object  
# df_past.info() # 일시 object 행정구 object
df_past['일시'] = df_past['일시'].astype('datetime64')
df_past2 = df_past[df_past['일시']>'2023-12-31']
df_final['월'] = df_final['일시'].dt.month
df_final['일'] = df_final['일시'].dt.day
df_past2['월'] = df_past2['일시'].dt.month
df_past2['일'] = df_past2['일시'].dt.day
df_past2
df_check = df_final.merge(df_past2, on=['월','일','행정구'])
df_check.drop('일시_y',axis=1,inplace=True)
```
