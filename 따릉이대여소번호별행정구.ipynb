```python
import pandas as pd
test = pd.read_csv('model_1/2022년_일별_대여소별_대여량.csv')
test.drop('Unnamed: 0',axis=1,inplace = True)
df_['대여소번호'] = df_['대여소번호'].astype(int)
df_2022 = test.merge(df_, on='대여소번호') # inner : 대여량이 없거나 활성화되지 않은 정류소까지 고려할 필요 X
df_2022 = df_2022.groupby(['행정구','월','일'],as_index=False)['count'].sum()
df_2022.to_csv('model_1/2022년_일별_행정구_대여량_총계.csv')
```
