```python
# 필요 라이브러리 호출
import dotenv
import os
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import time
import pandas as pd
import matplotlib.pyplot as plt
import requests
from bs4 import BeautifulSoup
import xml.etree.ElementTree as ET
import warnings
```

```python
spot_name = pd.read_excel('data/서울시 주요 120장소 목록.xlsx')['AREA_NM'].tolist()
```

```python
spot_list = [spot for spot in spot_name if spot.endswith('역')]
spot_list
```

# 역마다 행정동 주소 웹크롤링

```python
data_list = []
for spot in spot_list :
    if spot == '서울식물원·마곡나루역' :
        print(spot)
        data_list.append({'서울식물원·마곡나루역':'서울 강서구 마곡동'})
    elif spot == '신촌·이대역' :
        print(spot)
        data_list.append({'신촌·이대역':'서울 마포구 연남동'})
    else :
        keyword = spot.split()[0]
        url = 'https://search.naver.com/'
        driver = webdriver.Chrome()
        driver.get(url)
        elem = driver.find_element(By.NAME, 'query')
        time.sleep(2)
        elem.clear()
        elem.send_keys(keyword)
        elem.send_keys(Keys.RETURN) # Keys.ENTER : 엔터를 넣은 것, Keys.Return : submit을 하는 엔터
        time.sleep(8)
        elem2 = driver.find_element(By.CSS_SELECTOR, 'span.Pb4bU')
        print(elem2.text, '일거에요')
        data_list.append({spot : elem2.text})
        time.sleep(1)
        driver.close()
```

```python
data_list
```

```python
df_data = []
for data in data_list :
    df_data.append([list(data.keys())[0], data[list(data.keys())[0]]])
```

# 2차 가공 : 역별 행정동 -> 역별 행정구

```python
df = pd.read_csv('역별행정구.csv')
```

```python
df.head(2)
```

```python
df.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
df
```

```python
def gu_extract(row) :
    row = row.copy()
    row['행정구'] = row[1].split(' ')[1]
    return row
```

```python
df.loc[0]
```

```python
gu_extract(df.loc[0])
```

```python
df = df.apply(gu_extract, axis=1)
```

```python
df.columns = ['역명','행정동','행정구']
```

```python
df.drop('행정동',axis=1,inplace=True)
```
